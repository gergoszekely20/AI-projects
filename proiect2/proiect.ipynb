```python
train = train_data.copy()
```

```python
train_cleaned = train.drop(columns=["patient_id", "metastatic_first_novel_treatment_type", "metastatic_first_novel_treatment","patient_gender", "patient_race", "bmi"])
```

```python
train_cleaned['payer_type'] = train_cleaned['payer_type'].fillna('Unknown')
train_cleaned['patient_state'] = train_cleaned['patient_state'].fillna('Unknown')
train_cleaned['Region'] = train_cleaned['Region'].fillna('Unknown')
train_cleaned['Division'] = train_cleaned['Division'].fillna('Unknown')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Iterați peste fiecare coloană din DataFrame
for col in train_cleaned.columns:
    if train_cleaned[col].isnull().sum() > 0:
        imputer = IterativeImputer(max_iter=100)
        train_cleaned[col] = imputer.fit_transform(train_cleaned[[col]])
```

#### Forest Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt


def root_mean_squared_error(y_true, y_pred, squared=False):
    return sqrt(mean_squared_error(y_true, y_pred))

X = train_cleaned.drop('metastatic_diagnosis_period', axis=1)
y = train_cleaned['metastatic_diagnosis_period']
X_matrix= pd.get_dummies(X)
X_train_matrix, X_test_matrix, y_train, y_test = train_test_split(X_matrix, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_matrix, y_train)
y_predicted = model.predict(X_test_matrix)
rmse = root_mean_squared_error(y_test, y_predicted)
print('Forest Regression RMSE:', rmse)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt


def root_mean_squared_error(y_true, y_pred, squared=False):
    return sqrt(mean_squared_error(y_true, y_pred))

X = train_cleaned.drop('metastatic_diagnosis_period', axis=1)
y = train_cleaned['metastatic_diagnosis_period']
X_matrix= pd.get_dummies(X)
X_train_matrix, X_test_matrix, y_train, y_test = train_test_split(X_matrix, y, test_size=0.2, random_state=42)

model_linear = LinearRegression()
model_linear.fit(X_train_matrix, y_train)
y_predicted_linear = model_linear.predict(X_test_matrix)
rmse_linear = root_mean_squared_error(y_test, y_predicted_linear)
print('Linear Regression RMSE:', rmse_linear)
```

#### Ridge Regression

```python

from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from math import sqrt
import pandas as pd
from sklearn.model_selection import train_test_split

def root_mean_squared_error(y_true, y_pred, squared=False):
    return sqrt(mean_squared_error(y_true, y_pred))

X = train_cleaned.drop('metastatic_diagnosis_period', axis=1)
y = train_cleaned['metastatic_diagnosis_period']
X_matrix= pd.get_dummies(X)
X_train_matrix, X_test_matrix, y_train, y_test = train_test_split(X_matrix, y, test_size=0.2, random_state=42)

model_ridge = Ridge(alpha=1.0) 
model_ridge.fit(X_train_matrix, y_train)
y_predicted_ridge = model_ridge.predict(X_test_matrix)
rmse_ridge = root_mean_squared_error(y_test, y_predicted_ridge)
print('Ridge Regression RMSE:', rmse_ridge)
```

##### Concluzii
##### Regresia liniară cu regularizare Ridge are tendința de a fi mai ușor de interpretat și mai eficientă din punct de vedere computațional în comparație cu regresia aleatorie cu pădure (Random Forest), ceea ce poate fi important în funcție de cerințele și constrângerile specifice ale dumneavoastră.

### New feature and model training


```python
train = train_data.copy()
```

```python
cleand_train = train.drop(columns=["patient_id", "metastatic_first_novel_treatment_type", "metastatic_first_novel_treatment", "bmi", "patient_gender"])
```

```python
cleand_train['payer_type'] = cleand_train['payer_type'].fillna('Unknown')
cleand_train['patient_race'] = cleand_train['payer_type'].fillna('Unknown')
cleand_train['patient_state'] = cleand_train['patient_state'].fillna('Unknown')
cleand_train['Region'] = cleand_train['Region'].fillna('Unknown')
cleand_train['Division'] = cleand_train['Division'].fillna('Unknown')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Iterați peste fiecare coloană din DataFrame
for col in cleand_train.columns:
    if cleand_train[col].isnull().sum() > 0:
        imputer = IterativeImputer(max_iter=100)
        cleand_train[col] = imputer.fit_transform(cleand_train[[col]])
```

```python
cleand_train.isnull().sum()
```

```python
def categorize_age(age):
    if age < 30:
        return 'Under 30'
    elif age < 40:
        return '30s'
    elif age < 50:
        return '40s'
    elif age < 60:
        return '50s'
    elif age < 70:
        return '60s'
    else:
        return '70+'
cleand_train['age_category'] = cleand_train['patient_age'].apply(categorize_age)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
import pandas as pd

def root_mean_squared_error(y_true, y_pred):
    return sqrt(mean_squared_error(y_true, y_pred))

patient_features = ['patient_age', 'patient_race']
patient_payer_features = patient_features + ['payer_type']+['age_category']

X_encoded = pd.get_dummies(cleand_train[patient_features])
X_encoded_2 = pd.get_dummies(cleand_train[patient_payer_features])


X_train, X_test, y_train, y_test = train_test_split(X_encoded, cleand_train['metastatic_diagnosis_period'], test_size=0.2, random_state=42)
model_patient = RandomForestRegressor(n_estimators=100, random_state=42)
model_patient.fit(X_train, y_train)
y_pred_patient = model_patient.predict(X_test)



X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_encoded_2, cleand_train['metastatic_diagnosis_period'], test_size=0.2, random_state=42)
model_patient_payer = RandomForestRegressor(n_estimators=100, random_state=42)
model_patient_payer.fit(X_train_2, y_train_2)
y_pred_patient_payer = model_patient_payer.predict(X_test_2)


rmse_patient = root_mean_squared_error(y_test, y_pred_patient)
rmse_patient_payer = root_mean_squared_error(y_test_2, y_pred_patient_payer)

print("RMSE for Patient-Only Model:", rmse_patient)
print("RMSE for Patient + Payer Type Model:", rmse_patient_payer)
```

```python
from sklearn.ensemble import VotingRegressor

model_patient_payer = RandomForestRegressor(n_estimators=100, random_state=42)
model_patient = RandomForestRegressor(n_estimators=100, random_state=42)

ensemble = VotingRegressor(estimators=[('model_patient_payer', model_patient_payer), ('model_patient', model_patient)])

ensemble.fit(X_train, y_train)

y_pred_ensemble = ensemble.predict(X_test)

rmse_ensemble = root_mean_squared_error(y_test, y_pred_ensemble)

print("RMSE for Ensemble Model:", rmse_ensemble)
```

### Diagnosticat per cod cancer mamar
##### Se observa ca unele tipuri de cancer mamar sunt mai greu de diagnosticat inainte de 90 de zile decat altele. De exemplu, pentru cancerul mamar cu codul 1749, este mai dificil sa fie diagnosticat intr-un interval mai scurt. In schimb, pentru cancerul mamar cu codul C50911, este mai usor sa fie diagnosticat mai devreme. Acest lucru sugereaza ca timpul necesar pentru diagnosticare poate varia in functie de tipul de cancer mamar.

```python

filtered_data = train_data[train_data['breast_cancer_diagnosis_code'].isin(breast_cancer_count['breast_cancer_diagnosis_code'])]

association_analysis = pd.crosstab(filtered_data['breast_cancer_diagnosis_code'], filtered_data['DiagPeriodL90D'])

association_analysis['Diagnosticat <90 Zile (%)'] = (association_analysis[1] / association_analysis.sum(axis=1) * 100).round(2)

association_analysis

```

```python
test = test_data.copy()
```

```python
test_cleaned = test.drop(columns=["patient_id", "metastatic_first_novel_treatment_type", "metastatic_first_novel_treatment", "patient_race", "bmi", "Ozone", "PM25", "N02"])
```

```python
test_cleaned['payer_type'] = test_cleaned['payer_type'].fillna('Unknown')
test_cleaned['patient_state'] = test_cleaned['patient_state'].fillna('Unknown')
test_cleaned['Region'] = test_cleaned['Region'].fillna('Unknown')
test_cleaned['Division'] = test_cleaned['Division'].fillna('Unknown')
```

```python
# Importai clasa IterativeImputer din scikit-learn

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Iterai peste fiecare coloan din DataFrame
for col in test_cleaned.columns:
    if test_cleaned[col].isnull().sum() > 0:
        imputer = IterativeImputer(max_iter=100)
        test_cleaned[col] = imputer.fit_transform(test_cleaned[[col]])
```
